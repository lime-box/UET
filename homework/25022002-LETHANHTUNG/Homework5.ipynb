```python
A = []
m = int(input("Nhập số cột của ma trận:"))
n = int(input("Nhập số hàng của ma trận:"))
for i in range(n):
    A.append([])
    for k in range(m):
        t = input(f"Nhập phần tử {(i+1, k+1)} của ma trận (số nguyên > -1000 và < 10000):")
        if(len(t) < 4):
            t = " "*(4-len(t)) + t
        A[i].append(t)
for i in A:
    for k in i:
        print(k, end = " ")
    print()
```

<h1>W6A21

```python
A = input("Nhập 1 dòng các chuỗi (tách bởi dấu cách):").split()
B = {}
for i in range(len(A)):
    C = []
    if(A[i] != -1):
        C.append(i)
        for k in range(i+1, len(A)):
            if(A[k] == A[i]):
                C.append(k)
                A[k] = -1
        B[A[i]] = tuple(C)
        A[i] = -1
print(B)
```
