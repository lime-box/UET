<h1>W5A3</h1>

```python
def prime(n):
    if(n < 2 or n//1 != n):
        return False
    elif(n == 2):
        return True
    else:
        for i in range(2, int(n**.5) + 1):
            if(n%i == 0):
                return False
        return True

prime(97)
```

<h1>W5A4</h1>

```python
def perfect(n):
    if(n < 0 or n//1 != n):
        return False
    c = 0
    for i in range(1, n//2 + 1):
        if(n%i == 0):
            c += i
    if(n == c):
        return True
    else:
        return False

perfect(8128)
```

```python
def index(A, n):
    k = 0
    for i in A:
        if(i == n):
            return k
        k += 1
    return -1

index([5, 6, 9, 4, 6, 0], 6)
```

<h1>W5A6</h1>

```python
def factorial(n):
    if(n <= 0 or n//1 != n):
        return -1
    c = 1
    for i in range(1, n + 1):
        c *= i
    return c

factorial(10)
```

<h1>W5A7</h1>

```python
def round2(n):
    if((n - n//1)*100 - (n - n//1)*100//1 >= .5):
        return(n//1 + ((n - n//1)*100//1 + 1)/100)
    else:
        return(n//1 + ((n - n//1)*100//1)/100)

def calc(a, b, op):
    match op:
        case "+":
            return(round2(a + b))
        case "-":
            return(round2(a - b))
        case "*":
            return(round2(a * b))
        case "/":
            return(round2(a / b))

print(calc(2.896, 5.66, "/"))
```

<h1>W5A8</h1>

```python
def binary(n):
    c = [0]
    for i in range(n):
        if(c[-1] == 0):
            c[-1] = 1
        else:
            i = 1
            while i <= len(c):
                if(c[-i] == 0):
                    c[-i] = 1
                    break
                else:
                    c[-i] = 0
                    i += 1
            if(i > len(c)):
                c.insert(0, 1)
    return c

def hamming(a, b):
    if(b > a):
        a, b = b, a
    a, b = binary(a), binary(b)
    while(len(a) > len(b)):
        b.insert(0, 0)
    c = 0
    for i in range(len(a)):
        if(a[i] != b[i]):
            c += 1
    return c

hamming(56, 99)
```

```python
def sum_of_digits(n):
    c = 0
    for i in str(n):
        c += int(i)
    return c

sum_of_digits(7896)
```

```python
def count(a):
    n = []
    for i in a:
        n.append(i)
    N = []
    while(n != []):
        i = n[0]
        k = 0
        while(i in n):
            n.remove(i)
            k += 1
        if(N == []):
            N.append(k)
        else:
            for i in range(len(N)):
                if(k <= N[i]):
                    N.insert(i, k)
                    break
    return(N)

def mapping(a, b):
    if(count(a) == count(b)):
        return True
    else:
        return False

mapping("ajrdjaa", "hhejhss")
```
