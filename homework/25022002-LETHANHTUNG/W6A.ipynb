```python
A = tuple(map(int, input("Nhập 1 tuple số nguyên:").split()))
k = int(input("Nhập vị trí xoay:"))
B = []
for i in range(k, len(A)):
    B.append(A[i])
if(k > len(A)):
    for i in range(len(A)):
        B.append(A[i])
else:
    for i in range(k):
        B.append(A[i])
print(tuple(B))
```

```python
A = input("Nhập 1 dict các số:").split()
B = input("Nhập 1 dict các số:").split()
C = {}
for i in range(0, len(A), 2):
    C[A[i]] = float(A[i+1])
for i in range(0, len(B), 2):
    if(i in C):
        C[B[i]] += float(B[i+1])
    else:
        C[B[i]] = float(B[i+1])
print(C)
```

```python
A = list(map(int, input("Nhập 1 dòng các số nguyên:").split()))
A.sort()
B = []
k = int(input("Nhập 1 số nguyên dương:"))
for i in range(len(A)):
    for t in range(i+1, len(A)):
        if(A[i] + A[t] == k and (A[i], A[t]) not in B):
            B.append((A[i], A[t]))
print(B)
```
