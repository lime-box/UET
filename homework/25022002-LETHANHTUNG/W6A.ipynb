```python
n = input("Nhập các cặp key:value cách nhau bởi dấu cách:").split()
A = {}
for i in n:
    A[i.split(":")[0]] = int(i.split(":")[1])
print(A)
n = input("Nhập các cặp key:value cách nhau bởi dấu cách:").split()
B = {}
for i in n:
    B[i.split(":")[0]] = int(i.split(":")[1])
print(B)
C = []
D = []
for i in A:
    if(C == []):
        C.append(i)
        D.append(A[i])
    else:
        k = 0
        while(k < len(C)):
            if(i < C[k]):
                break
            k += 1
        C.insert(k, i)
        D.insert(k, A[i])
for i in B:
    if(C == []):
        C.append(i)
        D.append(B[i])
    else:
        k = 0
        while(k < len(C)):
            if(i == C[k]):
                D[k] += B[i]
                k = -1
                break
            if(i < C[k]):
                break
            k += 1
        if(k != -1):
            C.insert(k, i)
            D.insert(k, B[i])
E = {}
for i in range(len(C)):
    E[C[i]] = D[i]
print(E)
```

```python
A = list(map(int, input("Nhập 1 dòng các số nguyên:").split()))
B = []
k = int(input("Nhập 1 số nguyên dương:"))
for i in range(len(A)):
    for t in range(i+1, len(A)):
        if(A[i] + A[t] == k and (A[i], A[t]) not in B):
            z = 0
            for w in range(len(B)):
                if(A[i] < B[w][0]):
                    B.insert(w, (A[i], A[t]))
                    z = -1
                    break
            if(z == 0):
                B.append((A[i], A[t]))
print(B)
```
